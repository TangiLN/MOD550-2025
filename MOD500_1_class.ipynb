```python
import random
import matplotlib.pyplot as plt
import numpy as np
class Create_random_number:
    def __init__(self,number):
        """
        Initialize the class function
        Input : number size of the list we want to create
        """
        self.number=number
        print("Creation of an instance of : Create_random_number")
        
    def create_basic_list(self):
        """
        Create a list of ordered number
        Output: The orderred  list 
        """
        list_ordered=[]
        for i in range (self.number):
            list_ordered.append(i)
        return list_ordered
    
    def create_random_list(self):
        """
        Create a list of random number 
        Output: The random  list 
        """
        list_random=[]
        for i in range(self.number):
            list_random.append(random.randint(1,100))
        return list_random
    
    #Next task create a list of two dimension random number 

    def create_2d_list(self, x_size, y_size):
        """
        Create a 2D list of random number  
        Output: The random  list 
        """
        matrix_2d=[]
        for i in range(x_size):
            row=[]
            for j in range(y_size):
                row.append(random.randint(1,100))
            matrix_2d.append(row)
        return matrix_2d
    
x=np.linspace(1,1000,1000)
instance_creation_list=Create_random_number(10)
basic_list=instance_creation_list.create_basic_list()
# print(list_random)
print(instance_creation_list.create_random_list())
random_list=instance_creation_list.create_random_list()
random_list_2d=instance_creation_list.create_2d_list(100,2)
print(random_list_2d)
```

```python

class DataGenerator:
    def __init__(self,n_p=10):
        """Generate any data"""
        self.n_p=n_p

    def create_2d_list_random(self):
        """
        Create a 2D list of random number  with another method  
        Output: The random  list 
        """
        list_x=[]
        list_y=[]
        for i in range(self.n_p):
            list_x.append(random.random())
            list_y.append(random.random())
        self.list_xy=[[list_x],[list_y]]
    
    def create_2d_list_linear(self):
        """
        Create a 2D list of random number  with linear distribution  
        Output: The random  list 
        """
        list_x=[]
        list_y=[]
        for i in range(self.n_p):
            x=random.uniform(0,self.n_p) # same as random.randint but with float numbers 
            noise= random.gauss(0,self.n_p/10 )#random floating point number with gaussian distribution.
            y=2*x+noise
            list_x.append(x)
            list_y.append(y)
        self.list_xy=[[list_x],[list_y]]

    def plotter(self):
        """Generate Plots"""
        plt.scatter(*self.list_xy)
        plt.show()
        

datagenerator=DataGenerator(n_p=100)
datagenerator.create_2d_list_linear()
print(datagenerator.list_xy)
datagenerator.plotter()
```
