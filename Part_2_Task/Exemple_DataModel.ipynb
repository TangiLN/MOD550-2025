# Task 2_10 : Analize

- Linear Regression : 
 For the Linear Regression we supposed that the distribution is going to be linear with some noise, but if we give the function a random distribution it will falsificate the coeficient of the Line 
 Furthermore, if the distribution is non linear, the regression line will not ajust good because the function is going to have a big MSE. 
- Neural Network : 
 Impact of the number of layers 
 On the Neural Network, the distribution of data is important, if the variance is too small so data are too concentrate the NN is going to take more time to learn and could also not learn good information ("Trash"). 
 The more dimension there is in the dataset also the more the NN need layer, the more layer there is the more complex can be the dataset but if it's a simple distribution too many layers would be overfitting. 

- MSE : 
 The Mean Square Error is very sensitive to extreme values. The square in calculation increase drasticly when there is extreme points. 
 In order to have a good MSE we need to analyze the dataset and the extremes values of it, that way the distribution have a real impact on the MSE, a normalization of data can be useful to reduce the impact of the extreme data.  

- Clustering K-means :
 The Kmeans cluster are spheric and of a similar size, so what's important for this distribution is the variance of the data, if the distribution is too streched, the center of the cluster is going to be in the wrong place and if so the edges between each cluster is not going to make any sense. 
- Clustering GMM :
Gaussian Mixture Model is a cluster representation by using many Gaussian distribution, this distribution is quite flexible. 
In the case of a Uniform distribution this clustering methods could use too many Gaussian to approximate the distribution wich resume in a overfiting use of it.
